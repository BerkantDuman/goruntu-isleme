### Image rotation

```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
img_1 = plt.imread("cameraman.jpg")
```

```python
def display_tw_image(im_1, im_2):
    plt.figure(figsize=(10,10))
    plt.subplot(1,2,1)
    plt.imshow(im_1)
    plt.subplot(1,2,2)
    plt.imshow(im_2,)
    plt.show()

def rotate_by_i_j_swap(im_1):
    m,n,k = im_1.shape
    new_image = np.zeros((n,m,k), dtype="uint8")
    
    for i in range(m):
        for j in range(n):
            new_image[j,i] = im_1[i,j]
    return new_image
```

```python
img_2 = rotate_by_i_j_swap(img_1)
display_tw_image(img_1, img_2)
```

```python
def rotate_one_point_with_theta_counterclockwise(point, angle):
    theta = np.radians(angle)
    r = np.array(( (np.cos(theta), - np.sin(theta)),
                  (np.sin(theta), np.cos(theta)) ))
    
    v = np.array(point)
    
    return r.dot(v).astype(int)

def get_all_new_location(im_1, angle):
    m,n,k = im_1.shape
    new_loc_points = []
    
    for i in range(m):
        for j in range(n):
            new_loc_points.append(rotate_one_point_with_theta_counterclockwise([i,j], angle))
    return new_loc_points

def get_min_max(loc_points):
    min_x, min_y = loc_points[0][0], loc_points[0][1]
    max_x, max_y = loc_points[0][0], loc_points[0][1]
    
    s1 = len(loc_points)
    
    for s in range(s1):
        
        if min_x > loc_points[s][0]:
            min_x = loc_points[s][0]
        
        if max_x < loc_points[s][0]:
            max_x = loc_points[s][0]
            
        if min_y > loc_points[s][1]:
            min_y = loc_points[s][1]
            
        if max_y < loc_points[s][1]:
            max_y = loc_points[s][1]
    return min_x,min_y, max_x, max_y

def rotate_image(im_1, angle):
    m,n,k = im_1.shape
    new_loc_points = get_all_new_location(im_1, angle)
    min_x,min_y, max_x, max_y = get_min_max(new_loc_points)
    
    new_m = max_x - min_x +1
    new_n = max_y - min_y +1
    
    x_offset = 0-min_x
    y_offset = 0-min_y
    
    new_image = np.zeros((new_m, new_n, k), dtype="uint8")
    
    for i in range(m):
        for j in range(n):
            new_i, new_j = rotate_one_point_with_theta_counterclockwise([i,j],angle)
            new_image[new_i+x_offset, new_j+y_offset] = im_1[i,j]
    return new_image
```

```python
new_loc_points = get_all_new_location(img_1, 90)
min_x,min_y, max_x, max_y = get_min_max(new_loc_points)
rotate_one_point_with_theta_counterclockwise([0,1],90)
```

```python
img_3 = rotate_image(img_1, -80)
plt.imshow(img_3)
```
