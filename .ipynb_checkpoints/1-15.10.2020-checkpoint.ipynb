```python
import numpy as np
import os 
import matplotlib.pyplot as plt
```

```python
def get_jpg_files():
    path = os.getcwd()
    jpg_files = [f for f in os.listdir(path) if f.endswith('.jpg')]
    img1 = plt.imread(jpg_files[0])
    return img1
```

```python
def compare_types():
    list_1 = [1,2,3,4,5] #brodcast
    list_2 = [1,2,3,4,5]
    list_1 + list_2
    print(list_1 + list_2)
    list_1 = np.asarray([1,2,3,4,5]) #brodcast
    list_2 = np.asarray([1,2,3,4,5])
    
    print(list_1 + list_2 + 10)
compare_types()
```

```python
image = get_jpg_files()
image.shape
```
