```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib_inline
```

```python
img_1 = plt.imread("cameraman.jpg")
plt.imshow(img_1)
```

```python
img_1.ndim, img_1.shape
```

```python
def convert_rgb_gray_level(img):
    m =img.shape[0]
    n =img.shape[1]
    im_1 = np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            im_1[i,j]=get_distance(img[i,j,:])
    return im_1

def get_distance(v, w=[1/3,1/3,1/3]):
    a,b,c = v[0], v[1], v[2]
    w1,w2,w3 = w[0], w[1], w[2]
    d = ((a**2)*w1 +
         (b**2)*w2 +
         (c**2)*w3)**.5
    return int(d)
```

```python
def my_f_1(a, b):
    assert a >= 0; "Intensity possitive", "error intensity not positive"
    if(a <= (255-b)): return a+b
    else: return 255
```

```python
def my_f_2(a):
    #assert a >= 0; "Intensity possitive", "error intensity not positive"
    return int(255-a)
```

```python
img_2 = convert_rgb_gray_level(img_1)
plt.imshow(img_2, cmap="gray")
```

```python
m,n = img_2.shape
im_3 = np.zeros((m,n), dtype="uint8")
```

```python
for i in range(m):
    for j in range(n):
        intensity = img_2[i,j]
        increment=20
        im_3[i,j] = my_f_1(intensity, increment)
```

```python
plt.subplot(1,2,1)
plt.imshow(im_3, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_2, cmap="gray")
```

```python
m,n = img_2.shape
im_4 = np.zeros((m,n), dtype="uint8")
```

```python
for i in range(m):
    for j in range(n):
        intensity = img_2[i,j]
        im_4[i,j] = my_f_2(intensity)
```

```python
plt.subplot(1,3,1)
plt.imshow(im_4, cmap="gray")
plt.subplot(1,3,2)
plt.imshow(im_3, cmap="gray")
plt.subplot(1,3,3)
plt.imshow(img_2, cmap="gray")
```
